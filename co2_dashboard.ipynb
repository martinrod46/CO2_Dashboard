```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv')
```

```python
df
```

```python
df.columns
```

```python
df[df['country'] == 'World']
```

```python
#Fill NAs with 0s and create GDP per capita column 
## Note that we are going to use NumPy which is used for numerical computations. 
  #Provides support for multidimensional arrays and other mathematical functions to operate efficiently. ##

df = df.fillna(0)
df['gdp_per_capita'] = np.where(df['population']!=0, df['gdp']/df['population'],0) # <- This is basically an IF statement.. #
                                                                                #If 'Population" is different than 0 then #
                                                                            # GDP (Gross Domestic Product) is going to be divided by 'Population' #
                                                                            # and if the condition is not met then just 0 #
```

```python
# Make Dataframe Pipeline Interactive

idf=df.interactive()
```

```python
# Define Panel Widgets #

year_slider = pn.widgets.IntSlider(name='Year Slider', start=1750, end=2020, step=5, value=1850)
```

```python
year_slider
```

```python
# Radio buttons for Co2 measures

yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y Axis',
    options=['co2', 'co2_per_capita',],
    button_type='success'
    
)
```

```python
yaxis_co2
```

```python
continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America']

co2_pipeline = (
            idf[
                          (idf.year <= year_slider) & 
                          (idf.country.isin(continents))
             ]
             .groupby(['country','year'])[yaxis_co2].mean()
             .to_frame()
             .reset_index()
             .sort_values(by='year')
             .reset_index(drop=True)
) 
```

```python
co2_pipeline
```

```python
co2_plot = co2_pipeline.hvplot( x='year', by='country', y=yaxis_co2, line_width=2, title="CO2")
```

```python
co2_plot
```

Table - CO2 Emission Over Time By Continent

```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width')
```

```python
co2_table
```

CO2 vs GDP scatterplot

```python
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) & 
        ( ~ (idf.country.isin(continents)))
        ]
        .groupby(['country','year', 'gdp_per_capita'])['co2'].mean()
        .to_frame()
        .reset_index()
        .sort_values(by='year')
        .reset_index(drop=True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x='gdp_per_capita', y='co2', by='country', size=80, kind="scatter",
                                                                alpha=0.7, legend=False, height=500, width=500)
```

```python
co2_vs_gdp_scatterplot
```

Bar chart with CO2 sources by continent

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['coal_co2', 'oil_co2', 'gas_co2'],
    button_type='success'
)

continents_excl_world = ['Asia', 'Oceania' , 'Europe' , 'Africa', 'North America', 'South America', 'Antarctica']
co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year','country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind='bar', x='country', y=yaxis_co2_source, title='CO2 Source By Continent')
```

```python
co2_source_bar_plot
```

Creating Dashboard
